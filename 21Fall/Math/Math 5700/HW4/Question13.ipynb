# This is the cole for HW4 Question 13 bounded area

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-4,8,200)
y1=2*x-x**2
y2=-x
plt.plot(x,y1,'b-',label='$y=2x-x^2$')
plt.plot(x,y2,'g-',label='$x+y=0$')
plt.legend()
plt.xlabel('x')
plt.fill_between(x,y1,y2,'grey',alpha=0.5)
plt.title('Question 13 Bounded Graph')
plt.savefig('question13.pdf')
plt.show()
```
