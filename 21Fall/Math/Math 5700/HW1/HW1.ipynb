```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sympify, lambdify
from sympy.abc import x
```

```python
import numpy.ma as M  
from pylab import * 
xx = np.arange(-0.5,6.5,0.01) 
vals = np.sqrt(2*xx**2+1)/(3*xx-5)   
vals = M.array(vals) 
mvals = M.masked_where(xx==5/3, vals) 

subplot(121) 
plt.plot(xx, mvals, linewidth=3, color='red') 
xlim(-1,7) 
ylim(-3,10) 
plt.title('Question 3 Graph')
plt.savefig('question3.pdf')
```

```python
#Question10
x2=np.linspace(-1,10,200)
y1=np.exp(x2)-1
y2=np.exp(x2)
plt.figure()
plt.plot(x2,y1,'r--',label='$f\'(x)=e^x-1$')
plt.plot(x2,y2,'go',label='$f\'\'(x)=e^x$')
plt.title('Question 10')
plt.legend()
plt.xlabel('x')

plt.savefig('question10.pdf')
plt.show()
```
